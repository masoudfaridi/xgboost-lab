## XGBoost Parameters

| Parameter | Group | Purpose | Description |
|-----------|-------|---------|-------------|
| `objective` | Learning Objective | Defines the learning task | Specifies what problem the model is trying to solve. It determines the loss function used during training and how gradients and Hessians are computed. Examples include regression (`reg:squarederror`), binary classification (`binary:logistic`), multiclass classification (`multi:softprob`), ranking, and survival analysis. This is one of the first parameters that should be selected because it defines the entire optimization objective. |
| `base_score` | Learning Objective | Initial prediction | Sets the initial prediction (global bias) before any trees are built. XGBoost starts boosting from this value instead of zero. In most cases the default value works well, but it can be useful when prior knowledge about the target distribution is available. |
| `eval_metric` | Model Evaluation | Performance metric | Specifies how model performance is evaluated during training. Common metrics include RMSE, MAE, LogLoss, Error Rate, and AUC. Unlike `objective`, this parameter does **not** control optimization—it is only used to monitor performance and report training progress. |
| `early_stopping_rounds` | Model Evaluation | Stop unnecessary training | Stops training automatically if the evaluation metric does not improve after a specified number of boosting rounds. This reduces training time and helps prevent overfitting by avoiding unnecessary trees. |
| `booster` | Boosting Process | Select boosting algorithm | Chooses the underlying booster implementation. `gbtree` is the standard tree-based gradient boosting algorithm, `dart` introduces dropout for additional regularization, and `gblinear` trains a linear model instead of decision trees. |
| `learning_rate` (`eta`) | Boosting Process | Shrink tree contribution | Controls how much each newly built tree contributes to the final prediction. Smaller values slow down learning, usually requiring more trees but often improving generalization and reducing overfitting. |
| `n_estimators` | Boosting Process | Number of boosting rounds | Determines how many trees are added to the ensemble. Increasing this value generally reduces bias but increases training time and the risk of overfitting if not combined with a smaller learning rate. |
| `process_type` | Boosting Process | Continue or restart training | Determines whether training starts from scratch (`default`) or updates an existing model (`update`). Mainly used in advanced or incremental training workflows. |
| `num_parallel_tree` | Boosting Process | Parallel tree construction | Specifies how many trees are built in parallel during each boosting round. This parameter is mainly used for Random Forest mode inside XGBoost rather than standard Gradient Boosting. |
| `tree_method` | Tree Structure & Growth | Tree construction algorithm | Chooses how trees are built. Options such as `exact`, `approx`, and `hist` trade off speed, memory usage, and scalability. The histogram algorithm (`hist`) is the most commonly used for large datasets. |
| `grow_policy` | Tree Structure & Growth | Tree expansion strategy | Determines how nodes are expanded. `depthwise` grows trees level by level, while `lossguide` expands only the most promising leaves, often producing more efficient trees for large datasets. |
| `max_depth` | Tree Structure & Growth | Maximum tree depth | Limits how deep each tree can grow. Larger values allow more complex decision boundaries but also increase variance and the likelihood of overfitting. |
| `max_leaves` | Tree Structure & Growth | Maximum number of leaves | Limits the number of terminal nodes instead of limiting depth. This provides a more flexible way to control tree complexity, especially when using histogram-based tree construction. |
| `max_bin` | Tree Structure & Growth | Histogram resolution | Defines the maximum number of bins used to discretize continuous features when using histogram-based algorithms. Larger values improve split precision but require more memory and computation. |
| `gamma` | Tree Structure & Growth | Minimum split gain | Specifies the minimum reduction in loss required before creating a new split. Higher values make splitting more conservative, resulting in simpler trees and lower overfitting risk. |
| `min_child_weight` | Tree Structure & Growth | Minimum child node weight | Defines the minimum sum of Hessian values required for a child node. Larger values prevent small or unreliable splits, making the model more conservative and reducing overfitting. |
| `max_delta_step` | Tree Structure & Growth | Limit leaf updates | Restricts the maximum update applied to leaf weights during optimization. It is rarely needed for balanced datasets but can improve stability in highly imbalanced classification problems. |
| `subsample` | Sampling | Sample training rows | Specifies the fraction of training samples randomly selected before building each tree. Values below 1.0 introduce randomness, reduce variance, and often improve generalization. |
| `sampling_method` | Sampling | Row sampling strategy | Determines how training samples are selected. `uniform` performs random sampling, while `gradient_based` focuses more on observations with larger gradients, potentially improving efficiency. |
| `colsample_bytree` | Sampling | Sample features per tree | Randomly selects a subset of features before constructing each tree. This reduces correlation between trees and helps prevent overfitting. |
| `colsample_bylevel` | Sampling | Sample features per level | Randomly selects features independently at every tree level, introducing additional diversity during tree construction. |
| `colsample_bynode` | Sampling | Sample features per split | Randomly selects candidate features at every individual split. This introduces the highest level of randomness among the column sampling parameters. |
| `reg_alpha` | Regularization & Imbalance | L1 regularization | Applies an L1 penalty to leaf weights. It encourages sparse models by shrinking less important weights toward zero, which can also perform implicit feature selection. |
| `reg_lambda` | Regularization & Imbalance | L2 regularization | Applies an L2 penalty to leaf weights. Unlike L1, it reduces large weights smoothly, improving model stability and reducing overfitting. |
| `scale_pos_weight` | Regularization & Imbalance | Handle class imbalance | Assigns additional weight to the positive class during training. This is particularly useful for highly imbalanced classification problems where positive examples are much rarer than negative ones. |
| `enable_categorical` | Categorical Features | Native categorical support | Enables XGBoost's built-in handling of categorical variables, eliminating the need for manual one-hot encoding in many cases. |
| `max_cat_to_onehot` | Categorical Features | One-hot encoding threshold | Specifies the maximum number of categories for which one-hot encoding is automatically applied. Features with more categories use partition-based splitting instead. |
| `max_cat_threshold` | Categorical Features | Category split search | Limits the number of category partitions considered when searching for the best split. Larger values may improve split quality but increase computation time. |

## Learning Objective

### objective

| Property | Description |
|----------|-------------|
| **Parameter Name** | `objective` |
| **Purpose** | Defines the learning task and the loss function optimized during training. |
| **Stage in Algorithm** | Used throughout the entire boosting process to compute gradients and Hessians for each boosting iteration. |
| **Default Value** | `reg:squarederror` |
| **Allowed Values** | A predefined set of objectives provided by XGBoost (Regression, Classification, Ranking, Survival Analysis, etc.). |
| **Typical Choices** | `reg:squarederror`, `binary:logistic`, `multi:softprob` |
| **Recommended Starting Value** | Choose the objective that exactly matches the machine learning task. |
| **Most Related Parameter** | `eval_metric` |
| **Other Related Parameters** | `num_class`, `scale_pos_weight`, `base_score`, `tweedie_variance_power`, `quantile_alpha`, `expectile_alpha`. |
| **When to Change** | Whenever the learning task changes (e.g., Regression → Classification or Binary → Multiclass). |
| **Common Mistakes** | Using `multi:softmax` when class probabilities are required; forgetting to specify `num_class`; using regression objectives for classification problems. |
| **Tuning Priority** | ⭐⭐⭐⭐⭐ (Must be selected correctly before tuning any other hyperparameter.) |

| Objective | Task Type | Output | Typical Use Case |
|-----------|-----------|--------|------------------|
| `reg:squarederror` | Regression | Continuous value | Standard regression problems using Mean Squared Error (MSE). |
| `reg:squaredlogerror` | Regression | Continuous value | Regression where relative errors are more important than absolute errors. Labels must be greater than `-1`. |
| `reg:absoluteerror` | Regression | Continuous value | Robust regression when the dataset contains outliers. Optimizes Mean Absolute Error (MAE). |
| `reg:pseudohubererror` | Regression | Continuous value | Smooth approximation of MAE, balancing robustness and differentiability. |
| `reg:quantileerror` | Regression | Quantile prediction | Quantile regression for prediction intervals and uncertainty estimation. |
| `reg:expectileerror` | Regression | Expectile prediction | Expectile regression using asymmetric squared loss. |
| `reg:gamma` | Regression | Positive continuous value | Gamma-distributed targets such as insurance claim severity. |
| `reg:tweedie` | Regression | Positive continuous value | Tweedie-distributed targets (e.g., insurance total loss). |
| `reg:logistic` | Regression | Probability | Regression task with outputs constrained to the interval `[0,1]`. |
| `binary:logistic` | Binary Classification | Probability | Standard binary classification. |
| `binary:logitraw` | Binary Classification | Raw score | Binary classification without applying the logistic transformation. |
| `binary:hinge` | Binary Classification | Class label (0/1) | Binary classification requiring direct class predictions instead of probabilities. |
| `multi:softmax` | Multiclass Classification | Class label | Multiclass classification returning the predicted class index. |
| `multi:softprob` | Multiclass Classification | Probability vector | Multiclass classification returning class probabilities. |
| `count:poisson` | Count Regression | Count mean | Count data following a Poisson distribution. |
| `survival:cox` | Survival Analysis | Hazard ratio | Right-censored survival analysis using the Cox proportional hazards model. |
| `survival:aft` | Survival Analysis | Survival time | Accelerated Failure Time (AFT) survival model. |
| `rank:pairwise` | Learning to Rank | Ranking score | Pairwise ranking tasks. |
| `rank:ndcg` | Learning to Rank | Ranking score | Ranking optimized for NDCG. |
| `rank:map` | Learning to Rank | Ranking score | Ranking optimized for Mean Average Precision (MAP). |

### base_score

`base_score` defines the **initial prediction** of the model before the first tree is built. Unlike hyperparameters such as `learning_rate` or `max_depth`, it **does not control model complexity**. Instead, it determines the starting point of the boosting process.

In other words,

```text
Prediction before Tree #1 = base_score
```

The first tree learns to correct this initial prediction, and each subsequent tree further refines it.

#### Examples

##### Regression

If

```python
base_score = 100000
```

the model initially predicts `100000` for every training instance before any trees are built.

##### Binary Classification

With

```python
objective = "binary:logistic"
base_score = 0.5
```

the model initially assigns a **50% probability** to every instance, and the boosting process gradually updates these probabilities.

#### Default Behavior

Since **XGBoost 3.1.0**, `base_score` is **automatically estimated** from the training data. Therefore, manually setting this parameter is rarely necessary.

#### Effect on the Model

`base_score` does **not** affect the number of trees, tree depth, or model complexity. It only changes the **starting point** of optimization. With enough boosting rounds, different `base_score` values often converge to nearly the same final model; the main difference is in the early stages of training and, occasionally, the speed of convergence.

| Property | Description |
|----------|-------------|
| **Parameter Name** | `base_score` |
| **Purpose** | Specifies the initial prediction (global bias or intercept) before any boosting iteration begins. |
| **Stage in Algorithm** | Used before the first tree is constructed as the starting prediction for all training instances. |
| **Default Value** | Automatically estimated from the training data (since XGBoost 3.1.0). |
| **Previous Default** | `0.5` in older XGBoost versions. |
| **Allowed Values** | Any real number appropriate for the selected objective. |
| **Recommended Starting Value** | Leave it at the default (automatic estimation) unless there is a specific reason to override it. |
| **Most Related Parameter** | `objective` |
| **Other Related Parameters** | `base_margin` (overrides `base_score`), `eval_metric`. |
| **When to Increase** | Rarely. Only when reliable prior knowledge suggests a higher initial prediction. |
| **When to Decrease** | Rarely. Only when reliable prior knowledge suggests a lower initial prediction. |
| **Common Mistakes** | Assuming `base_score` controls model complexity or improves accuracy; manually changing it without a justified reason. |
| **Tuning Priority** | ⭐☆☆☆☆ (Very low. In most applications it should be left at its default value.) |

| Aspect | Increase `base_score` | Decrease `base_score` |
|--------|------------------------|------------------------|
| **Initial Prediction** | Starts training from a larger initial prediction. | Starts training from a smaller initial prediction. |
| **Effect on Model** | Shifts the initial prediction upward before any tree is built. | Shifts the initial prediction downward before any tree is built. |
| **Bias** | No significant effect after sufficient boosting rounds. | No significant effect after sufficient boosting rounds. |
| **Variance** | No significant effect. | No significant effect. |
| **Overfitting** | No direct effect. | No direct effect. |
| **Underfitting** | No direct effect. | No direct effect. |
| **Training Time** | May converge slightly faster if closer to the optimal initial prediction. | May require more corrections if farther from the optimal starting point. |
| **Prediction Time** | No effect. | No effect. |
| **Memory Usage** | No effect. | No effect. |
| **Training Stability** | Can slightly improve optimization if the initial value is well chosen. | Can slightly slow convergence if poorly chosen. |
| **Generalization** | Usually no noticeable effect. | Usually no noticeable effect. |

## Learning Parameters

## learning_rate

| Property | Description |
|----------|-------------|
| **Parameter Name** | `learning_rate` (`eta`) |
| **Purpose** | Controls the contribution (shrinkage) of each newly added tree before it is incorporated into the ensemble. |
| **Stage in Algorithm** | Applied **after a tree is trained** and **before** its predictions are added to the current ensemble. |
| **Default Value** | `0.3` |
| **Allowed Range** | $(0, 1]$ |
| **Typical Values** | `0.01`, `0.05`, `0.1`, `0.2`, `0.3` |
| **Recommended Starting Value** | `0.1` for most practical machine learning tasks. |
| **Quick Baseline** | `0.3` when fast experimentation is more important than maximum predictive performance. |
| **High-Accuracy Models** | `0.05` or `0.01`, typically combined with a large `n_estimators` and Early Stopping. |
| **Most Related Parameter** | `n_estimators` (strong inverse relationship). |
| **Other Related Parameters** | `max_depth`, `subsample`, `colsample_bytree`, `reg_alpha`, `reg_lambda`, `early_stopping_rounds`. |
| **When to Increase** | When the model clearly Underfits, training is too slow, or computational resources are limited. |
| **When to Decrease** | When the model Overfits, validation performance is unstable, or maximum generalization is desired. |
| **Common Mistakes** | Decreasing `learning_rate` without increasing `n_estimators`; using very large values (e.g., `>0.5`) on real-world datasets; tuning `learning_rate` independently of `n_estimators`. |
| **Tuning Priority** | ⭐⭐⭐⭐⭐ (One of the first hyperparameters that should be tuned.) |

| Aspect | Increase `learning_rate` | Decrease `learning_rate` |
|--------|---------------------------|---------------------------|
| **Effect on Model** | Each new tree has a larger contribution to the ensemble. The model learns more aggressively and converges in fewer boosting rounds. | Each new tree contributes less to the ensemble. The model learns more gradually and typically requires more boosting rounds. |
| **Bias** | ↓ Lower Bias, as the model adapts more quickly to the training data. | ↑ Higher Bias if `n_estimators` is fixed, because each tree makes only a small correction. Increasing `n_estimators` can compensate for this effect. |
| **Variance** | ↑ Higher Variance, since each tree has a stronger influence on the final model and the model becomes more sensitive to noise. | ↓ Lower Variance, because learning is distributed across many small updates, resulting in a more stable model. |
| **Overfitting** | Higher risk of Overfitting, especially when combined with deep trees or a large number of estimators. | Lower risk of Overfitting because the optimization process is more conservative. |
| **Underfitting** | Lower risk of Underfitting. However, excessively large values may overshoot the optimum and reduce generalization performance. | Higher risk of Underfitting if `n_estimators` is not increased accordingly. |
| **Training Time** | Faster training because fewer boosting iterations are usually required. | Slower training because more boosting iterations are generally needed. |
| **Prediction Time** | Usually faster inference because the final ensemble contains fewer trees. | Usually slower inference because the final ensemble typically contains more trees. |
| **Memory Usage** | Usually lower memory usage because fewer trees are stored. | Usually higher memory usage because more trees are stored in the model. |
| **Training Stability** | Lower optimization stability. Large updates may cause oscillation or unstable convergence. | Higher optimization stability due to smaller and smoother updates. |
| **Generalization** | May decrease if the learning rate is too large because the model becomes more sensitive to noise. | Often improves when combined with a sufficiently large `n_estimators`, leading to better robustness on unseen data. |

## n_estimators

| Property | Description |
|----------|-------------|
| **Parameter Name** | `n_estimators` |
| **Purpose** | Specifies the total number of boosting rounds (trees) to be built during training. |
| **Stage in Algorithm** | Determines how many sequential trees are added to the ensemble. Each iteration trains one new tree to correct the errors of the current model. |
| **Default Value** | `100` (Scikit-learn API)<br>`xgb.train()` uses `num_boost_round=10` by default if not specified. |
| **Allowed Range** | Positive integer (`n_estimators ≥ 1`). |
| **Typical Values** | `100`, `200`, `500`, `1000`, `3000` |
| **Recommended Starting Value** | `300–500` together with `learning_rate = 0.05–0.1` and Early Stopping. |
| **Quick Baseline** | `100` |
| **High-Accuracy Models** | `1000+` trees are common when using a very small `learning_rate` (e.g., `0.01–0.05`) and Early Stopping. |
| **Most Related Parameter** | `learning_rate` (strong inverse relationship). |
| **Other Related Parameters** | `early_stopping_rounds`, `max_depth`, `subsample`, `colsample_bytree`, `gamma`. |
| **When to Increase** | When the model Underfits, the validation metric is still improving, or a smaller `learning_rate` is used. |
| **When to Decrease** | When training becomes unnecessarily slow, the model Overfits, or Early Stopping indicates that additional trees provide no benefit. |
| **Common Mistakes** | Using a large `n_estimators` with a large `learning_rate`; selecting `n_estimators` without Early Stopping; reducing `learning_rate` without increasing `n_estimators`. |
| **Tuning Priority** | ⭐⭐⭐⭐⭐ (One of the first hyperparameters that should be tuned together with `learning_rate`.) |

| Aspect | Increase `n_estimators` | Decrease `n_estimators` |
|--------|--------------------------|--------------------------|
| **Effect on Model** | More trees are added to the ensemble, allowing the model to learn more complex patterns and reduce training error. | Fewer trees are built, resulting in a simpler model with limited learning capacity. |
| **Bias** | ↓ Lower Bias because the model has more opportunities to correct previous errors. | ↑ Higher Bias because the boosting process may stop before capturing the underlying relationships in the data. |
| **Variance** | ↑ Higher Variance, especially when combined with a large `learning_rate` or deep trees. | ↓ Lower Variance because the model complexity is reduced. |
| **Overfitting** | Higher risk of Overfitting if too many trees are added after the validation performance has stopped improving. | Lower risk of Overfitting because the model is less complex. |
| **Underfitting** | Lower risk of Underfitting since the model has sufficient boosting rounds to fit the data. | Higher risk of Underfitting because the model may terminate before learning important patterns. |
| **Training Time** | Longer training time due to more boosting iterations. | Shorter training time because fewer trees are trained. |
| **Prediction Time** | Slower inference since predictions from more trees must be aggregated. | Faster inference because fewer trees are evaluated. |
| **Memory Usage** | Higher memory usage because more trees are stored in the model. | Lower memory usage because the ensemble is smaller. |
| **Training Stability** | Generally stable, but excessive trees may continue fitting noise if early stopping is not used. | Stable but may produce an under-trained model. |
| **Generalization** | Can improve when paired with a small `learning_rate`; otherwise may decrease due to Overfitting. | May decrease if the model lacks sufficient capacity to learn the data. |
